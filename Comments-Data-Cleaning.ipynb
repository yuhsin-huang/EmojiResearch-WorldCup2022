```python
# scraped comment data is stored at Data/comments.csv
# clean comment data
import csv
import pandas as pd
comments_df = pd.read_csv("Data/comments.csv")
comments_df.head()
```

```python
list(comments_df.columns)
```

```python
print(len(comments_df.index))
```

```python
# Get comments from 11/20/2022-12/18/2022 (During the event)
comments_df['timestamp'] = pd.to_datetime(comments_df['timestamp'])  
start_date = '2022-11-20'
end_date = '2022-12-19'
mask = (comments_df['timestamp'] >= start_date) & (comments_df['timestamp'] <= end_date)
comments_df = comments_df.loc[mask]
```

```python
# number of rows after restricting the dates
print(len(comments_df.index))
```

```python
# sort the dataframe by timestamp
comments_df.sort_values(by=['timestamp'], inplace=True)
comments_df.head()
```

```python
# export dataframe to csv
import os  
comments_df.to_csv('/Data/out.csv')  
```
